```python
# 5. 버켓팅 적용
def bucket_conversations(df, bucket_size):
    buckets = {}
    for _, row in df.iterrows():
        text_length = len(row['cleaned_conversation'].split())
        bucket_id = (text_length // bucket_size) * bucket_size
        if bucket_id not in buckets:
            buckets[bucket_id] = []
        # 버켓에 대화 데이터를 추가할 때, 각 row 자체를 추가합니다.
        buckets[bucket_id].append(row.to_dict())  # 각 행을 딕셔너리 형태로 저장
    return buckets
```

```python
# 버켓 크기를 10으로 설정
bucket_size = 10
buckets = bucket_conversations(train_df, bucket_size)
```

```python
# 토크나이저를 정의한 후 각 버켓에 대해 패딩을 적용하는 함수
def tokenize_and_pad_all_buckets(buckets, tokenizer):
    tokenized_buckets = {}
    
    for bucket_range, bucket_data in buckets.items():
        if isinstance(bucket_data, list):  # 리스트인지 확인하여 토크나이징 처리
            print(f"Processing bucket {bucket_range} with {len(bucket_data)} conversations...")
            
            # 버켓 내 모든 문장 토크나이징 및 패딩
            texts = [row['cleaned_conversation'] for row in bucket_data]
            tokenized_data = tokenizer(texts, padding=True, truncation=True, max_length=128, return_tensors="pt")
            
            # 각 버켓의 토큰화된 데이터를 저장
            tokenized_buckets[bucket_range] = tokenized_data
        else:
            print(f"Skipping bucket {bucket_range} due to incompatible data type.")
    
    return tokenized_buckets
```

```python
# 각 버켓에 대해 토크나이징 및 패딩을 적용
tokenized_buckets = tokenize_and_pad_all_buckets(buckets, tokenizer)
```

```python
for bucket_range, tokenized_data in tokenized_buckets.items():
    print(f"Bucket {bucket_range}:")
    print(tokenized_data['input_ids'])
```
